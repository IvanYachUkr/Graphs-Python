```python
# (3 бали) Побудувати графiчний файл iз зображенням графа. Налаштувати розташування

# вершин графа так, щоб компоненти зв’язностi розташовувались послiдовно, а ребра мали якомо-
# га менше перетинiв. (Кiлькiсть ребер, яку достатньо вилучити для отримання плоского графа,

# мiнiмальна. За погодженням з викладачем у випадку, коли менше трьох перехресть отримати
# не вдається, може бути iнший критерiй, щоб отрималось не дуже захаращене перетинами ребер
# зображення.)
# Примiтка. Автоматизацiя розташування вершин не вимагається. Розташування вершин

# можна будувати повнiстю вручну або частково використовувати програмнi засоби (щоб зменши-
# ти кiлькiсть рутинної роботи з набору координат). При цьому заборонено використовувати

# бiблiотечнi засоби розташування вершин.

# Усi подальшi графiчнi зображення графу будувати за допомогою цього розташування вер-
# шин.
```

```python
# (3 бали) Для графа програмно знайти i в ЗРОЗУМIЛОМУ виглядi вивести в текстовому
# режимi для кожної його компонентi зв’язностi:
# кiлькiсть вершин та ребер, степенi та ексцентриситети вершин, радiус та дiаметр
# Create a subgraph SG based on a (possibly multigraph) G

for count, vertices in enumerate(nx.connected_components(G)):
    
    component = nx.Graph()
    for el in G.edges(vertices):
        component.add_edge(*el)
    print(f"MAIN INFORMATION ABOUT {count+1} CONNECTED COMPONENT: ")
    print("\n")
    
    print(f"Number of vertices of the {count+1} connected component is {nx.number_of_nodes(component)}. \n")
    print(f"Number of edges of the {count+1} connected component is {nx.number_of_edges(component)}. \n")

    degrees_list = [f"node: {pair[0]}, degree: {pair[1]} " for pair in nx.degree(component)]
    
    print(f"The degrees of the {count+1} connected component are: \n" )
    for string in  degrees_list:
        print(string)
    print("\n")
    eccentr_dict = dict(nx.eccentricity(component)) 
    eccentr_list = ([])
    print(f"The eccentricities of nodes of the {count+1} connected component are: \n")
    for key in eccentr_dict.keys():
        print(f"node: {key}, eccentricity: {eccentr_dict[key]}")
    print("\n")
    print(f"The radius of the {count+1} connected component is {nx.radius(component)}.")
    print(f"The diameter of the {count+1} connected component is {nx.diameter(component)}.")
    print("\n\n")
    
```

```python
# (3 бали) Для кожної нетривiальної компоненти зв’язностi програмно знайти хоча б один
# дiаметр (простий ланцюг, вiдстань мiж кiнцями якого дорiвнює дiаметру графа). Побудувати
# графiчне зображення, де ребра та вершини, що входять до знайдених дiаметрiв, вiдмiчено
# кольором. (Вручну пiдсвiчувати його ребра кольорами не можна :) )

def find_diameter(component):
    """
    find the first and the last nodes of diameter 
    """
    d = nx.diameter(component)
    path = dict(nx.all_pairs_shortest_path(component))
    keys = path.keys();

    for key in path.keys():
        for k in path.keys():
            if k != key and len(path[key][k]) == d+1:
                return path[key][k]
            
# def _define_nodes_color(component, diameter):
#     """
#     create a list of colores for nodes
#     "red" for nodes in diameter
#     "white" for other nodes
#     """
#     node_colors = []
#     for node in nx.nodes(component):
#         if node in diameter:
#             node_colors.append("red")
#         else:
#             node_colors.append("white")
#     return node_colors
def _define_nodes_color(component, diameter):
    """
    create a list of colores for nodes
    "red" for nodes in diameter
    "white" for other nodes
    """

    for node in nx.nodes(component):
        if node in diameter:
            yield "red"
        else:
            yield "white"


def _define_edges_color(component, diameter):
    """
    create a list of colores for edges
    "red" for edges in diameter
    "white" for other edges
    """
    num_of_nodes = len(diameter) 
    
    edges_shortest_path = [(diameter[i],diameter[i+1])
                           for i in range (num_of_nodes-1)]
    
    #edge_colors = []
            
    for edge in nx.edges(component):
        
        if edge in edges_shortest_path :

            yield "red"

        else:
            yield "black"
            
    #return edge_colors

def draw_diameter_CG(component):
    """
    draw diameter for connected component
    """
      
    diameter = find_diameter(component)
    options["node_color"] =[color for color in _define_nodes_color(component, diameter)]
    options["edge_color"] = [color for color in _define_edges_color(component, diameter)]
    
    nx.draw_networkx(component, pos= pos,  **options)
    
    options["node_color"]= "white" 
    options["edge_color"] = "black"
    

for vertices in nx.connected_components(G):
    
        component = nx.Graph()
        for el in G.edges(vertices):
            component.add_edge(*el)    
        draw_diameter_CG(component)     

```

```python
# 6) (3 бали) Програмно побудувати глибинний кiстяковий лiс графа. Побудувати зображення
# графа, на якому ребра знайденого лiсу видiлено кольором.
# (Якщо лiс знайдено програмно, то вручну пiдсвiчувати його ребра кольорами не можна :) )
T = nx.Graph() 
for edge in nx.dfs_edges(G, source=None, depth_limit=None):
    T.add_edge(*edge)
options["edge_color"] = "red"
nx.draw_networkx(T, pos= pos,  **options)
nx.draw_networkx_edges(G, width=1.0, pos =pos, edgelist = [edge for edge in G.edges if edge not in T.edges], edge_color="black")
```
